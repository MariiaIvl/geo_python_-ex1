```python

first_name= "Мария"
second_name= "Ивлева"
ice_cream_rating = 10
sleep_rating = 10

happiness_rating = (ice_cream_rating + sleep_rating) / 2
happiness_percentage = happiness_rating / 10 * 100

print(f"Меня зовут {first_name}, и я ставлю мороженому {ice_cream_rating} баллов из 10!")
print(f"Я {first_name} {second_name}, и мой рейтинг удовольствия от сна — {sleep_rating} из 10!")
print(f"Исходя из вышеперечисленных факторов, мой рейтинг счастья составляет {happiness_rating} из 10, или {happiness_percentage:.1f}%!")
```
